```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\soumy\Downloads\archive (2)\netflix_titles.csv")   #Dataset load 
df.head()
```

```python
df.info()
df.isnull().sum()    #checking missing values and the structure 
```

```python
# Data Cleaning
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')   # converting date column
```

```python
sns.countplot(x='type', data=df)  counting how many are Movies and TV Shows on Netflix
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')   #counting which country produce most shows and movies
plt.title("Top 10 Countries")
plt.show()
```

```python
df['year_added'] = df['date_added'].dt.year   #check how many shows and movies added over the years
df['year_added'].value_counts().sort_index().plot()
plt.title("Content Added Over Years")
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar') #it checks which is most common (Action, Comedy,Horror)
plt.title("Top Genres")
plt.show()
```
